```python
from sklearn.model_selection import train_test_split

# Assuming you have a DataFrame 'df' with features and target variable
X = credit_df.drop(columns=['Class'])  # Features (independent variables)
y = credit_df['Class']  # Target variable (dependent variable

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
y_test
```

```python
#Since we have to transform an array into z score we are using the below statemnet
from scipy import stats
y_train
```

```python
#Best way to transform train and test data set separately
#X_train = zscore(X_train,axis=1,ddof=1) #transforming training set to z transformation, to ensure all the data are in same scale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
X_train_scaled
```

```python
X_test_scaled
```

# Importing Support Vector Classifier

```python
from sklearn.svm import SVC

#Instantiating the model(means storing the property of SVC package in variable svc)
svc = SVC()

#Fitting the model using the classifier
svc.fit(X_train_scaled, y_train)
```

# Determining accuracy of the training and test set

```python
print("Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled,y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

# Trying to improve the model by using C=1000 technique, however accuracy remains the same

```python
svc = SVC(C=1000)
svc.fit(X_train_scaled, y_train)
m = svc.predict(X_test_scaled)
print(m)
print("Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

# Using Confusion matrix to understand the Precision and Recall for 0 and 1(non-fraud and fraudulent)

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,m))
```
